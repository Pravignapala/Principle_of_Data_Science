```python
import pandas as pd

# Replace 'your_file.csv' with the actual path to your CSV file
df = pd.read_csv('https://raw.githubusercontent.com/Pravignapala/Principle_of_Data_Science/refs/heads/main/Data.csv')

# If you want to remove leading/trailing whitespace from each column
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Print the DataFrame
print(df)
```

```python
df.columns = df.columns.str.replace(' ', '')
print(df['Frailty'])
```

```python
# Convert 'Frailty' to numerical (0/1)
df['Frailty'] = df['Frailty'].map({'N': 0, 'Y': 1})


# Save the cleaned data
df.to_csv('frailty_data_cleaned.csv', index=False)

print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Gripstrength', hue='Frailty', data=df)
plt.title('Age vs. Grip Strength with Frailty')
plt.xlabel('Age')
plt.ylabel('Grip Strength')
plt.show()
```

```python
print("## Conclusion")
print("Based on the analysis of the Frailty dataset, we can observe the following:")
print("- **Correlation:** Age and Gripstrength are negatively correlated with Frailty. This means, that as age and grip strength increase, the probability of frailty decreases. ")
print("- **Visual Analysis:** The scatter plot between Age and Grip Strength with Frailty highlighted, further shows this pattern. Individuals with higher grip strength and lower age seem less prone to frailty.")
print("These findings suggest that age and grip strength might be potential indicators or contributing factors to frailty.")
```
